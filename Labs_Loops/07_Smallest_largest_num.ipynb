Smallest and largest numbers in a list
Max Score: 10 | Difficulty: medium

Write a program that reads a list of integers into a list as long as the integers are greater than zero, then outputs the smallest and largest integers in the list.

Ex: If the input is:

10
5
3
21
2
-6

the output is:

2 and 21

```python
i = 0
l1 = [int(x) for x in input("Enter numbers: ").split() if int(x) > 0]

smallest = l1[0]
largest = l1[0]

for num in l1:
    if num < smallest:
        smallest = num
    if num > largest:
        largest = num

print("Smallest number:", smallest)
print("Largest number:", largest)
```
