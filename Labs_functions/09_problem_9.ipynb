Define a function called exact_change that takes the total change amount in cents and calculates the change using the fewest coins. The coin types are pennies, nickels, dimes, and quarters. Then write a main program that reads the total change amount as an integer input, calls exact_change(), and outputs the change, one coin type per line. Use singular and plural coin names as appropriate, like 1 penny vs. 2 pennies. Output "no change" if the input is 0 or less.

Ex: If the input is:

0
(or less), the output is:

no change
Ex: If the input is:

45
the output is:

2 dimes
1 quarter
Your program must define and call the following function. The function exact_change() should return num_pennies, num_nickels, num_dimes, and num_quarters.
def exact_change(user_total)

dime = 10 cent
nickel = 5 cent
penny = 1 cent
quarter = 25 cent


```python


def exact_change(user_total):
    quarter = user_total // 25
    user_total = user_total - (quarter * 25)

    dime = user_total // 10 
    user_total = user_total - (dime * 10)

    nickel = user_total // 5
    user_total = user_total - (nickel * 5)

    penny = user_total 

    return quarter, dime, nickel, penny





user_total = int(input())

if user_total == 0:
    print("No change")
else:
    quarter, dime, nickel, penny = exact_change(user_total)

    if quarter > 0:
        print(f"{quarter} Quarter{'s' if quarter > 1 else ''}")
    if dime > 0:
        print(f"{dime} Dime{'s' if dime > 1 else ''}")
    if nickel > 0:
        print(f"{nickel} Nickel{'s' if nickel > 1 else ''}")
    if penny > 0:
        print(f"{penny} Penny{'s' if penny > 1 else ''}")
        
```
