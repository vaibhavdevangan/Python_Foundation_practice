Davy's auto shop services

(2) Prompt the user for two services from the menu. (2 pts)

Select first service:

Car wax

(3) Output an invoice for the services selected. Output the cost for each service and the total cost. (3 pts)

Davy's auto shop invoice

Service 2: Car wax, $12

Total: $47

(4) Extend the program to allow the user to enter a dash (-), which indicates no service. (3 pts)

Ex:

Davy's auto shop services

Oil change -- $35
Tire rotation -- $19
Car wash -- $7
Car wax -- $12
Select first service:
Tire rotation
Select second service: _

Davy's auto shop invoice

Service 1: Tire rotation, $19
Service 2: No service

Total: $19

```python
print('''
Davy's auto shop services

Oil change -- $35
Tire rotation -- $19
Car wash -- $7
Car wax -- $12''')

services = { "oil change" : "$35",
             "tire rotation" : "$19",
             "car wash" : "$7",
             "car wax" : "$12",
             }


first_service = input("Select the first service :").strip().lower()
second_service = input("Select the first service:").strip().lower()


print(f'''
Davy's auto shop invoice
Service 1: {first_service}, {services[first_service]}

Service 2: {second_service}, {services[second_service]}

''')
total = int(services[first_service]) + int(services[second_service])
print(f'Total: ${total}')









# automobile_service = input("Enter desired auto service :").strip().lower()

# if automobile_service in services:
#     print(f"You entered: {automobile_service}")
#     print(f"Cost of oil change: {services[automobile_service]}")
```
